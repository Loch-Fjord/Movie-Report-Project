```python
# reading parquet
df = pd.read_parquet("/content/drive/My Drive/Parquets/by_votes.gzip")
```

```python
df.head(5)
```

```python
# making a column user_score which is vote_average*10 rounded to the nearest 1s
df["user_score"] = round(df["vote_average"]*10)

# making it an int instead of a float
df.user_score = df.user_score.astype(int)

df.head(5)
```

```python
# making a column release year which is the release year taken from release_date
df["release_year"] = df["release_date"].str[:4]

# making it into a number
df.release_year = pd.to_numeric(df.release_year)

df.head(5)
```
