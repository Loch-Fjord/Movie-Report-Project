```python
def column_of_lists_to_frequency_and_percent(df,column_name):
    '''
    makes a df with columns count and % of Top 800 Movies
    from a df with columns that contain lists of items
    '''
    # makes column into list
    i_list = df[column_name].to_list()
    # flattens out 3d list into 2d list
    i_list = [num for row in i_list for num in row]
    # makes list into pandas series
    i_list = pd.Series(i_list)
    # value_counts
    thing_counts = i_list.value_counts()

    # gets # of rows (how many movies?)
    total_rows = len(df[df[column_name].notnull()])

    # converts into dataframe
    thing_counts = thing_counts.to_frame(name='count')

    # makes a column that shows the % of top 800 movies thing applies to
    thing_counts["% of Top 800 Movies"] = (thing_counts["count"]/total_rows)*100

    return thing_counts
```

```python
genre_counts = column_of_lists_to_frequency_and_percent(df,"genre_names")
keyword_counts = column_of_lists_to_frequency_and_percent(df,"keywords")
actor_counts = column_of_lists_to_frequency_and_percent(df,"actors")
```
